###### line graph below depicts the moving average(50,300,500) in 'Close' for APPLE in all times.
###### Now, we can see the overall trend of closing stock prices. Basically the moving averages smoothens the rough edges of the graph and provides better understanding over the time.

```python
AAPL['MA50'] = AAPL['Close'].rolling(50).mean()
AAPL['MA300'] = AAPL['Close'].rolling(300).mean()
AAPL['MA500'] = AAPL['Close'].rolling(500).mean()
AAPL[['Close','MA300','MA50','MA500']].plot(figsize=(20,10),grid='True',legend='True', rot=0)
```

###### line graph below depicts the return percentage change in all times for APPLE. We see a massive around 1600% return from the day this data was received 2006. We can calculate the return percentage change by subtracting the points on the graph for the two time period to be calculated.

```python
a=AAPL['Close'].to_frame()
apple_return_percent_change = a.apply(lambda x: ((x - x[0])/x[0]))*100
apple_return_percent_change.plot(figsize=(20,10),grid='True',legend='True', rot=0)
```

###### line graph below depicts the return rate from 2015 onwards for APPLE. We see a 55% return.

```python
stocks_apple = pd.DataFrame({"APPLE": AAPL['1/1/2015':]["Close"]})
apple_return_percent_change = stocks_apple.apply(lambda x: ((x - x[0])/x[0])*100)
apple_return_percent_change.plot(figsize=(20,10),grid='True',legend='True', rot=0)
```

```python
#  Pandas describe function gives descriptive statistics, default assumes you are dealing with a sample
# Pandas also deals with missing values by ommitting them 
apple_return_percent_change.describe()
```

```python
from scipy import stats
stats.describe(apple_return_percent_change[1:])
```

```python
#  for comparision generate random numbers that follow normal distribution
x = np.random.normal(505.457787,417.007942, 3018)
stats.describe(x)
```

```python
#  use scipy stats to test the kurtosis, the null hypothesis is that the sample is drawn from a population 
# where the underlying kurtosis is that of a normally distributed variable
print("x: ", stats.kurtosistest(x))
print("amzn: ", stats.kurtosistest(apple_return_percent_change[1:]) )
```
